Import libraries

```python
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
```

Load and preprocess an image to make a prediction with the model

```python
model_path = "/content/drive/MyDrive/repos/keras-distinguish-own-cat-from-others-model/models/model1.keras"

model = tf.keras.models.load_model(model_path)

IMG_SIZE = (180, 180)
img_path = "drive/MyDrive/datasets/dataset_fla_other/preprocessed_images/flaekli/flaekli_147.jpg"

img = keras.utils.load_img(img_path, target_size=IMG_SIZE)
plt.imshow(img)
plt.axis("off")
plt.show()

img_array = keras.utils.img_to_array(img)
img_array = img_array / 255.0
img_array = tf.expand_dims(img_array, axis=0)

predictions = model.predict(img_array)

score = float(predictions[0])
print(f"This image is {100 * (1 - score):.2f}% Flaekli and {100 * score:.2f}% other cat.")
```
